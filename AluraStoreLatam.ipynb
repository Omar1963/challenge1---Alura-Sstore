#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Concatenar los dataframes de las 4 tiendas
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Calcular el total de facturación por producto y formatear a 3 decimales
todas_tiendas['Total Facturado'] = (todas_tiendas['Precio'] * todas_tiendas['Cantidad de cuotas']).round(3)

# Calcular la "Cantidad de Unidades" según el criterio definido y redondear al entero más cercano
# Criterio: Total Facturado / (Precio + Costo de envio), redondeado al entero más cercano.
# Nota: Este cálculo no representa la cantidad real de unidades vendidas si 'Cantidad de cuotas' es el número de pagos.
todas_tiendas['Cantidad de Unidades'] = (todas_tiendas['Total Facturado'] / (todas_tiendas['Precio'] + todas_tiendas['Costo de envío'] + 1e-9)).round().astype(int)


# Mostrar las primeras filas del dataframe con la nueva columna
display(todas_tiendas.head())
```

# Task
Analiza las ventas por producto en cada tienda para identificar el producto estrella y el producto de menor impacto en cada una.

## Calcular ventas por producto y por tienda

### Subtask:
Agrupar los datos por tienda y producto, y calcular el total facturado para cada combinación.


**Reasoning**:
I need to concatenate the individual store dataframes, add a column to identify the store, calculate the total billed amount, and then group by store and product to sum the total billed.



```python
import pandas as pd

# Concatenate the dataframes of the 4 stores, adding a store identifier
todas_tiendas = pd.concat([
    tienda.assign(Tienda='Tienda 1'),
    tienda2.assign(Tienda='Tienda 2'),
    tienda3.assign(Tienda='Tienda 3'),
    tienda4.assign(Tienda='Tienda 4')
], ignore_index=True)

# Calculate the total billing per row
todas_tiendas['Total Facturado'] = todas_tiendas['Precio'] * todas_tiendas['Cantidad de cuotas']

# Group by store and product and sum the total billed
ventas_por_tienda_producto = todas_tiendas.groupby(['Tienda', 'Producto'])['Total Facturado'].sum()

display(ventas_por_tienda_producto.head())
```

## Identificar producto estrella por tienda

### Subtask:
Para cada tienda, encontrar el producto con el mayor total facturado.


**Reasoning**:
To find the product with the highest total billed amount for each store, I will use the `idxmax()` method on the grouped series `ventas_por_tienda_producto` at the 'Tienda' level.



```python
productos_estrella_por_tienda = ventas_por_tienda_producto.groupby(level='Tienda').idxmax()
display(productos_estrella_por_tienda)
```

## Identificar producto de menor impacto por tienda

### Subtask:
Para cada tienda, encontrar el producto con el menor total facturado.


**Reasoning**:
Group the ventas_por_tienda_producto Series by the 'Tienda' level and find the product with the minimum total billed amount for each store.



```python
productos_menor_impacto_por_tienda = ventas_por_tienda_producto.groupby(level='Tienda').idxmin()
display(productos_menor_impacto_por_tienda)
```

## Presentar los resultados

### Subtask:
Mostrar los productos estrella y los de menor impacto para cada tienda, quizás junto con sus ventas totales correspondientes. Opcionalmente, relacionar estos hallazgos con el promedio general de ventas de cada tienda.


**Reasoning**:
Print the headings and iterate through the identified star and lowest impact products for each store, printing their names and total sales.



```python
print("--- Productos Estrella (Mayor Venta) por Tienda ---")
for store, product in productos_estrella_por_tienda.items():
    total_sales = ventas_por_tienda_producto.loc[store, product]
    print(f"{store}: {product} (Total Facturado: {total_sales:,.3f})")

print("\n--- Productos de Menor Impacto (Menor Venta) por Tienda ---")
for store, product in productos_menor_impacto_por_tienda.items():
    total_sales = ventas_por_tienda_producto.loc[store, product]
    print(f"{store}: {product} (Total Facturado: {total_sales:,.3f})")

# Optional: Calculate and print the overall average total facturado
average_total_facturado = ventas_por_tienda_producto.mean()
print(f"\nPromedio general de Total Facturado por Producto (todas las tiendas): {average_total_facturado:,.3f}")
```

## Calcular el ranking de ventas por producto dentro de cada tienda

### Subtask:
Dentro de cada tienda, asignar un rango a cada producto basado en su 'Total Facturado'.


**Reasoning**:
I will group the data by store and product, sum the total billed amount, and then rank the products within each store based on the total billed amount in descending order.



```python
# Group by store and product and sum the total billed
ventas_por_tienda_producto = todas_tiendas.groupby(['Tienda', 'Producto'])['Total Facturado'].sum()

# Rank products within each store based on Total Facturado in descending order
ranking_productos_por_tienda = ventas_por_tienda_producto.groupby(level='Tienda').rank(ascending=False)

display(ranking_productos_por_tienda.head())
```

## Preparar los datos para el gráfico

### Subtask:
Crear un DataFrame adecuado para el gráfico de puntos, con columnas para la tienda, el producto, las ventas y el ranking.


**Reasoning**:
Convert the Series `ventas_por_tienda_producto` and `ranking_productos_por_tienda` into DataFrames and merge them to create a single DataFrame with store, product, sales, and rank.



```python
# Convert ventas_por_tienda_producto Series to DataFrame and reset index
ventas_df = ventas_por_tienda_producto.reset_index(name='Total Facturado')

# Convert ranking_productos_por_tienda Series to DataFrame and reset index
ranking_df = ranking_productos_por_tienda.reset_index(name='Ranking')

# Merge the two DataFrames
producto_ranking_df = pd.merge(ventas_df, ranking_df, on=['Tienda', 'Producto'])

# Display the head of the new DataFrame
display(producto_ranking_df.head())
```

## Generar el gráfico de puntos

### Subtask:
Crear un scatter plot donde el eje X represente las tiendas, el eje Y represente el ranking de ventas (invertido para que el rango 1 esté arriba), y los puntos sean los productos, quizás diferenciados por color o tamaño según las ventas o la categoría.


**Reasoning**:
Generate a scatter plot to visualize the product sales ranking by store, using the created `producto_ranking_df` DataFrame.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Merge producto_ranking_df with todas_tiendas to get 'Categoría del Producto' for coloring
# Select only the necessary columns from todas_tiendas and drop duplicates before merging
producto_categoria = todas_tiendas[['Tienda', 'Producto', 'Categoría del Producto']].drop_duplicates()

# Merge the ranking dataframe with the category information
producto_ranking_df = pd.merge(producto_ranking_df, producto_categoria, on=['Tienda', 'Producto'], how='left')

plt.figure(figsize=(15, 10))
sns.scatterplot(data=producto_ranking_df,
                x='Tienda',
                y='Ranking',
                size='Total Facturado',  # Size of points based on total sales
                hue='Categoría del Producto', # Color based on product category
                alpha=0.7)

plt.title('Ranking de Ventas por Producto y Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ranking de Ventas')
plt.gca().invert_yaxis()  # Invert y-axis to show rank 1 at the top
plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
